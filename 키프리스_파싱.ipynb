# 요청 쿼리 작성

```python
url = "http://kportal.kipris.or.kr/kportal/resulta.do"
```

```python
headers = {
    "Referer": "http://kportal.kipris.or.kr/kportal/search/total_search.do"
    , "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/96.0.4664.110 Safari/537.36"
} # 윈도우 환경에서 유저가 접속했다고 페이지에 알려줌 
```

```python
search = input('검색어를 입력해 주십시오: ')

payload ={
    
"next": "patentList",
"FROM": "SEARCH",
"searchInTransKorToEng": "N",
"searchInTransEngToKor": "N",
"row": 30,
"page": int(input('크롤링할 페이지 수를 입력해 주십시오: ')),
"queryText": search,
"expression": search,
"strstat": ""

}
```

# 쿼리 요청

```python
res = requests.post(url, headers= headers ,  data = payload)
```

# XML 파싱

```python
import xml.etree.ElementTree as ET
from xml.etree.ElementTree import Element, SubElement, ElementTree, dump
import xml.etree.ElementTree as elemTree

tree = elemTree.fromstring(res.text)
```

# 데이터프레임화

```python
art = [] # 리스트를 저장할 list 초기화한다
iter_element = tree.iter(tag="article") # article태그 iterator를 가져옵니다
for element in iter_element: # article 태그를 순회합니다
    animal = {} # 각 저장할 dict 초기화한다
    animal['TTL>'] = element.find("TTL").text 
    animal['ABV>'] = element.find("ABV").text 
    art.append(animal) # 리스트에 정보를 저장합니다
print(art) # 결과를 출력한다
```

# 데이터프레임 출력

# csv로 저장 

```python
import pandas as pd

df_art = pd.DataFrame(art)
df_art.columns = ["특허이름", "서지정보"]

print(df_art)

df_art.to_csv("sample.csv",mode="w",encoding='euc-kr')
```
