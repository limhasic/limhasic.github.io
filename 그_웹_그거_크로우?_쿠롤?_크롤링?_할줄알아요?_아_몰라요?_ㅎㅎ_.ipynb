# 정규식

```python
import re # 정규식 라이브러리 
```

```python
# 정규식 
p = re.compile("ca.e") 
# . (ca.e) : 하나의 문자를 의미 > care, cafe, case  | caffe 
# ^ (^de) : 문자열의 시작   > desk, destination | fade
# $ (se$) : 문자열의 끝 > base, case | face 
```

```python
m = p.match("case")
print(m.group()) # 매치가 안되면 오류 메세지
```

```python
def print_match(m):
 
  #m = p.match(text)

  if m: 
    print(m.group())
  else:
    print("매칭되지 않음")
```

```python
print_match("case")
```

```python
print_match("good")
```

```python
m = p.search("good care") # 주어진 문자열 중에 일치하는게 있는지 확인
```

```python
print_match(m)
```

```python
def print_match(m):
 
  #m = p.match(text)

  if m: 
    print(m.group()) 
    print(m.string)
    print(m.start())
    print(m.end())
    print(m.span())

  else:
    print("매칭되지 않음")
```

```python
print_match(m)
```
